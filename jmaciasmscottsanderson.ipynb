```
m = [[9, 8, 6],
     [5, 4, 3],
     [2, 1, 0],]

n = [[5, 6, 7],
     [3, 2, 7]]
```

```
def multiplicar_matrices(m, n):
    filas = len(m)
    columnas = len(n[0])
    resultado = [[0 for col in range(columnas)] for fil in range(filas)]
    
    for i in range(filas):
        for j in range(columnas):
            for k in range(len(n)):
                resultado[i][j] += m[i][k] * n[k][j]
    return resultado
```

```
%%time

multiplicar_matrices(m, n)
```

```
def python_punto(Xa, Ya):
    return sum(X * Y for X, Y in zip(Xa, Ya))
```

```
%%fortran
subroutine fortran_punto(Xa, Ya, result)
    double precision, intent(in) :: Xa(:)
    double precision, intent(in) :: Ya(:)
    double precision, intent(out) :: result
    
    result = sum(Xa * Ya)
end
```

```
lista_datos = [float(j) for j in range(999999)]
arreglo_datos = np.array(lista_datos)
```

```
%%time
python_punto(lista_datos, lista_datos)
```

```
%%time
fortran_punto(arreglo_datos, arreglo_datos)
```

```
from av.aguacates import read_avocadata

aguacates = read_avocadata('2014', '2016')
aguacates.head()
```

```
aguacates.dtypes
```

```
sthin = aguacates[aguacates.Variety == 'HASS']
sthin.groupby(['Date', 'Region'])['Weighted Avg Price'].mean().unstack().ffill().plot();
```

```
def orsd(gr):

    if len(gr.columns) != 2:
        return pd.Series(index=gr.index, data=0.0)
    
    is_organic = gr.columns.get_level_values('Organic').values.astype(bool)
    organics = gr.loc[:, is_organic].squeeze()
    non_organics = gr.loc[:, ~is_organic].squeeze()
    diff = organics - non_organics
    return diff

def orsdr(dsf):
    """Hey my text!
    """
    return (
        dsf
        .set_index(['Date', 'Region', 'Organic'])
         ['Weighted Avg Price']
        .unstack(level=['Region', 'Organic'])
        .ffill()
        .groupby(level='Region', axis=1)
        .apply(orsd)
    )
```

```
orsdr(sthin).plot();
plt.gca().set_title("My Title");
plt.legend(bbox_to_anchor=(1, 1));
```

```
scotion = orsdr(sthin).corr()
scotion
```

```
import seaborn as sns
lad = sns.clustermap(scotion, annot=True)
fig = lad.fig
axes = fig.axes
ax = axes[3]
ax.set_xticklabels(ax.get_xticklabels(), rotation=78);
```
